### 🧠 Experiment 1: Uniform Cost Search (UCS) for Global Supply Chain Route Optimization

```python
from queue import PriorityQueue

graph = {
    'Dhaka': {'Chittagong': 4, 'Kolkata': 6},
    'Chittagong': {'Singapore': 5},
    'Kolkata': {'Dubai': 10, 'Istanbul': 12},
    'Singapore': {'Dubai': 7},
    'Dubai': {'Berlin': 8},
    'Istanbul': {'Berlin': 5},
    'Berlin': {}
}

def ucs(start, goal):
    pq = PriorityQueue()
    pq.put((0, start, [start]))
    visited = set()

    while not pq.empty():
        cost, node, path = pq.get()
        if node == goal:
            return path, cost
        if node in visited:
            continue
        visited.add(node)
        for neighbor, edge_cost in graph[node].items():
            pq.put((cost + edge_cost, neighbor, path + [neighbor]))
    return None, float('inf')

path, total_cost = ucs('Dhaka', 'Berlin')
print("Optimal Shipping Route:", path)
print("Total Transportation Cost:", total_cost)
```

### 💹 Experiment 2: Greedy Search for Dynamic Ad Campaign Optimization

```python
ads = {
    'Ad1': {'cost': 150, 'roi': 0.8},
    'Ad2': {'cost': 250, 'roi': 1.2},
    'Ad3': {'cost': 300, 'roi': 0.9},
    'Ad4': {'cost': 200, 'roi': 1.0},
    'Ad5': {'cost': 100, 'roi': 0.7}
}

def greedy_ads(ads, budget):
    sorted_ads = sorted(ads.items(), key=lambda x: x[1]['roi'], reverse=True)
    selected, total_cost = [], 0
    for ad, data in sorted_ads:
        if total_cost + data['cost'] <= budget:
            selected.append(ad)
            total_cost += data['cost']
    return selected, total_cost

chosen, spent = greedy_ads(ads, 500)
print("Selected Ads:", chosen)
print("Total Budget Used:", spent)
```

### 🚁 Experiment 3: A Search for Autonomous Drone Delivery Navigation*

```python
from queue import PriorityQueue
import math

grid = {
    'S': {'A': 4, 'B': 3},
    'A': {'C': 6, 'D': 5},
    'B': {'D': 2},
    'C': {'G': 2},
    'D': {'G': 3},
    'G': {}
}

heuristic = {'S': 8, 'A': 6, 'B': 7, 'C': 2, 'D': 3, 'G': 0}

def a_star(start, goal):
    pq = PriorityQueue()
    pq.put((heuristic[start], 0, start, [start]))
    while not pq.empty():
        f, g, node, path = pq.get()
        if node == goal:
            return path, g
        for neighbor, cost in grid[node].items():
            pq.put((g + cost + heuristic[neighbor], g + cost, neighbor, path + [neighbor]))
    return None, float('inf')

path, cost = a_star('S', 'G')
print("Drone Path:", path)
print("Total Cost:", cost)
```

### 🏪 Experiment 4: Hill Climbing for Dynamic Warehouse Space Optimization

```python
import random

products = ['A','B','C','D','E']
demand = [80, 60, 50, 30, 10]
layout = random.sample(products, len(products))

def efficiency(layout):
    return sum(demand[i] * (len(products)-idx) for idx, i in enumerate(range(len(layout))))

def hill_climb(layout):
    current = layout[:]
    while True:
        improved = False
        for i in range(len(current)):
            for j in range(i+1, len(current)):
                neighbor = current[:]
                neighbor[i], neighbor[j] = neighbor[j], neighbor[i]
                if efficiency(neighbor) > efficiency(current):
                    current = neighbor
                    improved = True
        if not improved:
            return current

best = hill_climb(layout)
print("Optimized Layout:", best)
print("Efficiency Score:", efficiency(best))
```

### 👷 Experiment 5: Genetic Algorithm for Hospital Shift Scheduling

```python
import random

employees = ['E1','E2','E3','E4','E5']
shifts = ['Morning','Evening','Night']
POP_SIZE = 10
MUT_RATE = 0.3

def create_schedule():
    return [random.choice(shifts) for _ in employees]

def fitness(schedule):
    score = 100
    if schedule.count('Night') > 2: score -= 10
    if len(set(schedule)) < 3: score -= 15
    return score

def crossover(p1, p2):
    point = random.randint(1, len(employees)-1)
    return p1[:point] + p2[point:]

def mutate(schedule):
    if random.random() < MUT_RATE:
        i = random.randint(0, len(schedule)-1)
        schedule[i] = random.choice(shifts)
    return schedule

def genetic_algorithm():
    population = [create_schedule() for _ in range(POP_SIZE)]
    for _ in range(50):
        population = sorted(population, key=fitness, reverse=True)
        new_gen = population[:2]
        while len(new_gen) < POP_SIZE:
            p1, p2 = random.sample(population[:5], 2)
            child = mutate(crossover(p1, p2))
            new_gen.append(child)
        population = new_gen
    return population[0]

best = genetic_algorithm()
print("Best Schedule:", best)
print("Fitness Score:", fitness(best))
```
